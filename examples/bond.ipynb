# 导入库

```python
import efinance as ef
```

# 使用例子

### 获取可转债日 K 线数据

```python
# 可转债代码（以 东财转3 为例）
bond_code = '123111'
ef.bond.get_quote_history(bond_code)
```

### 获取可转债 5 分钟 K 线数据

```python
# 可转债代码（以 东财转3 为例）
bond_code = '123111'
# 5 分钟
frequency = 5
ef.bond.get_quote_history(bond_code, klt=frequency)
```

### 全部可转债发行信息

```python
ef.bond.get_all_base_info()
```

### 可转债整体行情

```python
ef.bond.get_realtime_quotes()
```
